```python
import requests
from bs4 import BeautifulSoup
import re
from io import BytesIO
import pandas as pd

def scrape_csv_links(page_url):
    response = requests.get(page_url)
    response.raise_for_status()
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    csv_links = []
    for link in soup.find_all('a', href=True):
        href = link['href']
        label = link.text.strip()
        if href.endswith('.csv') and 'deprivation' in label.lower():
            csv_links.append((label, href))
    
    return csv_links

deprivation_scores = pd.DataFrame()

dfs = []
for year in [2010, 2015, 2019]:
    url = f"https://www.gov.uk/government/statistics/english-indices-of-deprivation-{year}"
    csv_files = scrape_csv_links(url)
    for name, href in csv_files:
        print(f"{name} -> {href}")
        
        response = requests.get(href)
        response.raise_for_status()
        
        dfs.append(pd.read_csv(BytesIO(response.content)))
        
for df in dfs:
    print(df.columns)
```

```python
RENAME_MAP = {
    # LSOA codes
    'LSOA CODE': 'lsoa_code',
    'LSOA code (2011)': 'lsoa_code',

    # IMD Rank
    'RANK OF IMD SCORE (where 1 is most deprived)': 'imd_rank',
    'Index of Multiple Deprivation (IMD) Rank (where 1 is most deprived)': 'imd_rank',

    # Income
    'RANK OF INCOME SCORE (where 1 is most deprived)': 'income_rank',
    'Income Rank (where 1 is most deprived)': 'income_rank',

    # Employment
    'RANK OF EMPLOYMENT SCORE (where 1 is most deprived)': 'employment_rank',
    'Employment Rank (where 1 is most deprived)': 'employment_rank',

    # Health
    'RANK OF HEALTH DEPRIVATION AND DISABILITY SCORE (where 1 is most deprived)': 'health_rank',
    'Health Deprivation and Disability Rank (where 1 is most deprived)': 'health_rank',

    # Education
    'RANK OF EDUCATION SKILLS AND TRAINING SCORE (where 1 is most deprived)': 'education_rank',
    'Education, Skills and Training Rank (where 1 is most deprived)': 'education_rank',

    # Crime
    'RANK OF CRIME SCORE (where 1 is most deprived)': 'crime_rank',
    'Crime Rank (where 1 is most deprived)': 'crime_rank',

    # Housing Barriers
    'RANK OF BARRIERS TO HOUSING AND SERVICES SCORE (where 1 is most deprived)': 'barriers_rank',
    'Barriers to Housing and Services Rank (where 1 is most deprived)': 'barriers_rank',

    # Living Environment
    'RANK OF LIVING ENVIRONMENT SCORE (where 1 is most deprived)': 'living_environment_rank',
    'Living Environment Rank (where 1 is most deprived)': 'living_environment_rank',

    # Indoors / Outdoors
    'Indoors Sub-domain Rank (where 1 is most deprived)': 'indoors_rank',
    'Outdoors Sub-domain Rank (where 1 is most deprived)': 'outdoors_rank',

    # Geographical / Wider barriers
    'Geographical Barriers Sub-domain Rank (where 1 is most deprived)': 'geographical_barriers_rank',
    'Wider Barriers Sub-domain Rank (where 1 is most deprived)': 'wider_barriers_rank',

    # Education subdomains
    'Children/Young People Sub-domain Rank (where 1 is most deprived)': 'children_rank',
    'Children and Young People Sub-domain Rank (where 1 is most deprived)': 'children_rank',
    'Skills Sub-domain Rank (where 1 is most deprived)': 'skills_rank',
    'Adult Skills Sub-domain Rank (where 1 is most deprived)': 'skills_rank',

    # Income Deprivation Indices
    'Rank of IDACI (where 1 is most deprived)': 'idaci_rank',
    'Income Deprivation Affecting Children Index (IDACI) Rank (where 1 is most deprived)': 'idaci_rank',
    'Rank of IDAOPI (where 1 is most deprived)': 'idaopi_rank',
    'Income Deprivation Affecting Older People (IDAOPI) Rank (where 1 is most deprived)': 'idaopi_rank',
}

columns = []

years = [2010, 2015, 2019]

for i in range(len(dfs)):
    dfs[i] = dfs[i].loc[:, dfs[i].columns.isin(RENAME_MAP)]
    dfs[i].rename(columns=RENAME_MAP, inplace=True)
    dfs[i]['year'] = years[i]
    columns.append(set(dfs[i].columns))
    
all(c == columns[0] for c in columns)
```

```python
merged = pd.concat([df for df in dfs], axis=0).reset_index()

import numpy as np
from scipy.stats import norm

def gaussianize_ranks_by_year(df, rank_columns, year_col='year'):
    df_transformed = df.copy()

    for year, group in df.groupby(year_col):
        idx = group.index  # indices to update in the original df
        print(idx)
        for col in rank_columns:
            u = group[col] / (len(group[col]) + 1)
            z = norm.ppf(u)
            # Apply inverse standard normal CDF ()
            df_transformed.loc[idx, col] = z

    return df_transformed

rank_cols = merged.columns[merged.columns.str.contains('rank')]
merged = gaussianize_ranks_by_year(merged, rank_cols)
```
